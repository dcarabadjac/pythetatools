```python
from pythetatools.config_visualisation import *
from pythetatools.config_fc import param_values_FC

from analysis_scripts import feldman_cousins
from matplotlib import pyplot as plt

import os
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('retina')
plt.style.use(t2k_style)

%load_ext autoreload
%autoreload 2
```

## Set output directory

```python
outdir_path = f'{outputs_dir}/plots/{CONFIG.dir_ver}/FC/'
os.makedirs(outdir_path, exist_ok=True)
outdir_files_path = f'{outputs_dir}/files/{CONFIG.dir_ver}/FC/'
os.makedirs(outdir_path, exist_ok=True)
```

# Set the parameter for Feldman-Cousins

```python
param_name = 'delta'
```

# Set input templates containing the FC fits

```python
basename_directory = (
    f"MargTemplates_LucileThesis_fit_FC{suffix[param_name]}_{{true_param_str}}_{{true_mh}}"
    f"_FullSyst_wRC_etheta_ptheta_{param_name}"
)
base_dir = f"{inputs_dir}/{CONFIG.dir_ver}/FC/{basename_directory}"
file_pattern = f"{base_dir}/marg*.root"
```

## Plot $\Delta \chi^2$ distributions and write the critical values in numpy and ROOT files

```python
labels = ['68.27%', '90.00%', '95.45%', '99.73%']

for true_param_str in param_values_FC[param_name]:
    for true_mh in [0, 1]:
        true_param_val = float(true_param_str)
        file_pattern = f"{base_dir}/marg*.root".format(true_param_str=true_param_str, true_mh=true_mh)
        feldman_cousins.feldman_cousins(file_pattern, true_param_val, true_mh, levels, outdir_path, outdir_files_path)
```

# Plot the critical values

```python
for true_mh in [0, 1]:
    feldman_cousins.plot_crit_val(param_name, true_mh, CONFIG, outdir_path, outdir_files_path)
  
```
